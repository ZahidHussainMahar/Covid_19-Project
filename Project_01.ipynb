# Q10

```python
raw_data= df.groupby('Country_Region')['Confirmed','Deaths','Recovered','Active'].sum().reset_index()
raw= raw_data[raw_data['Deaths']>150]

plt.figure(figsize=(12,5))

plt.plot(raw.Country_Region,raw.Confirmed, label='Confirmed')
plt.plot(raw.Country_Region,raw.Deaths, label='Deaths')
plt.plot(raw.Country_Region,raw.Recovered, label='Recovered')
plt.plot(raw.Country_Region,raw.Active, label='Active')

plt.title('Covid_19 Deaths > 150')
tick_mark=[i for i in range(len(raw['Country_Region']))] 
plt.xticks( rotation='vertical')

plt.legend()
plt.show()
```

# Q11

```python
usa=df[df['Country_Region']=='US']
usa=usa[usa.sum(axis=1)>0]
raw_data=usa.groupby('Province_State')['Deaths'].sum().reset_index()
raw_data= raw_data[raw_data['Deaths']>0]

plt.figure(figsize=(12,5))
plt.bar(raw_data.Province_State, raw_data.Deaths, label='Deaths')

plt.title('US Province_State wise Covid_19 Deaths Cases')
x=[i for i in range(len(raw_data['Province_State']))]
plt.xticks(x, rotation='vertical')

plt.legend()
plt.show()
```

# Q12

```python
us=df[df['Country_Region']=='US']
raw_data=us.groupby('Province_State')['Active'].sum().reset_index()
raw_data=raw_data[raw_data['Active']>0]

plt.figure(figsize=(12,5))

plt.bar(raw_data.Province_State,raw_data.Active, edgecolor='Green', label='Active')

plt.title('US Province_State wise Covid_19 Active Cases')
x=[i for i in range(len(raw_data['Province_State']))]
plt.xticks(x, rotation='vertical')

plt.legend()
plt.show()
```

# Q13

```python
pak=df[df['Country_Region']=='Pakistan']
pak=pak.groupby('Province_State')['Confirmed','Deaths','Recovered','Active'].sum().reset_index()

plt.figure(figsize=(15,5))

plt.bar(pak.Province_State,pak.Confirmed, label='Confirmed')
plt.bar(pak.Province_State,pak.Deaths, label='Deaths')
plt.bar(pak.Province_State,pak.Recovered, label='Recovered')
plt.bar(pak.Province_State,pak.Active, label='Active')

plt.title('Pakistan Province wise Covid_19 Cases')
plt.xlabel('Provinces')
plt.ylabel('Cases')
plt.legend()
plt.show()
```

# Q14

```python
world=df.groupby('Last_Update')['Confirmed','Deaths','Recovered','Active'].sum().reset_index()

plt.figure(figsize=(12,5))

plt.plot(world.Last_Update, world.Confirmed, label='Confirmed' )
plt.plot(world.Last_Update, world.Deaths, label='Deaths' )
plt.plot(world.Last_Update, world.Recovered, label='Recovered' )
plt.plot(world.Last_Update, world.Active, label='Active' )

plt.title('World Wide Covid_19 Cases Over the Time')
plt.xlabel('Last Update')
plt.ylabel('Cases')

plt.legend()
plt.show()
```
